```python
# fill all null values in interval datatype column to its mean value using imputer function

from pyspark.ml.feature import Imputer

impu = Imputer(inputCols=['win_by_runs','win_by_wickets'], outputCols=['{}_imputed'.format(c) for c in ['win_by_runs','win_by_wickets']]).setStrategy('mean')

impu.fit(spark_matches).transform(spark_matches).show()
```

```python
# find all winning teams who won by run >= 20

spark_matches.filter('win_by_runs >= 20').show()
```

```python
# find all winning team names who won by run >= 20

spark_matches.filter('win_by_runs >= 20').select('winner').show()
```

```python
# inverse condition operation

spark_matches.filter(~(spark_matches['win_by_runs']>=20)).select('winner').show()
```

# Machine Learning Models

```python
from pyspark.sql import SparkSession
import pandas as pd

spark = SparkSession.builder.getOrCreate()

df_pd = pd.read_excel('/content/delivery.xlsx')

# Convert the Pandas DataFrame to a Spark DataFrame
delivery_df = spark.createDataFrame(df_pd)
```

```python
delivery_df['Dist','Time'].show()
```

```python
delivery_df.printSchema()
```

```python
# consider Dist as independent variable and Time as independent variable

from pyspark.ml.feature import VectorAssembler

feature_assem = VectorAssembler(inputCols=['Dist'], outputCol='feature')
```

```python
# transform the current data to new column created above

new_df = feature_assem.transform(delivery_df)

new_df.show()
```

```python
final_data=new_df['feature','Time']
final_data.show()
```

```python
# Conduct Linear Regression

from pyspark.ml.regression import LinearRegression

train_data, test_data = final_data.randomSplit([0.60, 0.40], seed = 123)
```

```python
train_data.count()
```

```python
test_data.count()
```

```python
regressor = LinearRegression(featuresCol='feature',labelCol='Time')
regressor = regressor.fit(train_data)
```

```python
regressor.coefficients
```

```python
regressor.intercept
```

```python
# Predictions

pred_data = regressor.evaluate(test_data)

pred_data.predictions.show()
```

```python
# Calculate Absolute error, MSE

pred_data.meanAbsoluteError, pred_data.meanSquaredError
```
